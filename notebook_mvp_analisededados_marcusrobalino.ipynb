Foram aplicadas transforma√ß√µes para converter vari√°veis categ√≥ricas em formato num√©rico:

- **Gender:** Convertido para `Gender_Encoded` (Female = 0, Male = 1)  
- **Affects_Academic_Performance:** Convertido para `Affects_Academic_Performance_Encoded` (No = 0, Yes = 1)

Essas transforma√ß√µes s√£o essenciais para futuras an√°lises que envolvam m√©todos estat√≠sticos ou de aprendizado de m√°quina.


```python
from sklearn.preprocessing import LabelEncoder

# Inicializando o LabelEncoder
le = LabelEncoder()

# Codificando a coluna 'Gender'
df['Gender_Encoded'] = le.fit_transform(df['Gender'])

# Codificando a coluna 'Affects_Academic_Performance' (Yes/No ‚Üí 1/0)
df['Affects_Academic_Performance_Encoded'] = le.fit_transform(df['Affects_Academic_Performance'])

# Exibindo as novas colunas para valida√ß√£o
df[['Gender', 'Gender_Encoded', 'Affects_Academic_Performance', 'Affects_Academic_Performance_Encoded']].head()
```

###üìå 3. Transforma√ß√£o de Vari√°veis Num√©ricas

### Normaliza√ß√£o e Padroniza√ß√£o de Vari√°veis Num√©ricas:

Nesta etapa, aplicamos duas t√©cnicas de transforma√ß√£o em tr√™s vari√°veis num√©ricas do dataset:

- **Vari√°veis transformadas:**  
`Avg_Daily_Usage_Hours`, `Sleep_Hours_Per_Night`, `Mental_Health_Score`

---

#### ‚úîÔ∏è StandardScaler (Padroniza√ß√£o):
Transforma os dados para uma escala com m√©dia 0 e desvio padr√£o 1.  
**Objetivo:** Reduzir os efeitos de vari√°veis com magnitudes muito diferentes, preservando as distribui√ß√µes relativas.

**Vari√°veis padronizadas:**  
- `Avg_Daily_Usage_Hours_StandardScaled`  
- `Sleep_Hours_Per_Night_StandardScaled`  
- `Mental_Health_Score_StandardScaled`

---

#### ‚úîÔ∏è MinMaxScaler (Normaliza√ß√£o):
Transforma os dados para um intervalo fixo (0 a 1), preservando a proporcionalidade entre os valores originais.  
**Objetivo:** Tornar os dados compar√°veis em escalas fixas.

**Vari√°veis normalizadas:**  
- `Avg_Daily_Usage_Hours_MinMaxScaled`  
- `Sleep_Hours_Per_Night_MinMaxScaled`  
- `Mental_Health_Score_MinMaxScaled`

---

Ao final deste processo, o dataset conta agora com **seis novas colunas**, prontas para futuras an√°lises quantitativas mais avan√ßadas.

Essa prepara√ß√£o √© fundamental para garantir a qualidade e a integridade dos resultados nas pr√≥ximas etapas.


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Criando c√≥pias das colunas num√©ricas que vamos transformar
numerical_features = ['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score']

# Aplicando StandardScaler

scaler_standard = StandardScaler()
scaled_standard = scaler_standard.fit_transform(df[numerical_features])

# Criando um DataFrame com os resultados do StandardScaler
df_standard = pd.DataFrame(scaled_standard, columns=[col + '_StandardScaled' for col in numerical_features])

# Concatenando ao DataFrame original
df = pd.concat([df, df_standard], axis=1)

# Aplicando MinMaxScaler

scaler_minmax = MinMaxScaler()
scaled_minmax = scaler_minmax.fit_transform(df[numerical_features])

# Criando um DataFrame com os resultados do MinMaxScaler
df_minmax = pd.DataFrame(scaled_minmax, columns=[col + '_MinMaxScaled' for col in numerical_features])

# Concatenando ao DataFrame original
df = pd.concat([df, df_minmax], axis=1)

# Exibindo o resultado final

df.head()
```

###‚úÖ Hip√≥tese 1 (H1) ‚Äì Tempo de Uso x Sa√∫de Mental

## An√°lise da Hip√≥tese 1 (H1)

**Hip√≥tese:**

```python
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')

sns.regplot(
    x='Avg_Daily_Usage_Hours',
    y='Mental_Health_Score',
    data=df,
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'red'}
)

plt.title('H1 ‚Äì Rela√ß√£o entre Tempo de Uso e Sa√∫de Mental', fontsize=16, fontweight='bold')
plt.xlabel('Tempo M√©dio de Uso Di√°rio (Horas)', fontsize=12)
plt.ylabel('Mental Health Score', fontsize=12)

# Grade horizontal mais leve
plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()
```

## An√°lise da Hip√≥tese 3 (H3)

**Hip√≥tese:**  
Estudantes em relacionamentos afetivos est√°veis apresentam mais conflitos relacionados ao uso de redes sociais do que estudantes solteiros.

**Objetivo da An√°lise:**  
Verificar se existe diferen√ßa significativa no n√∫mero de conflitos sociais gerados pelo uso das redes sociais (`Conflicts_Over_Social_Media`) entre estudantes com diferentes status de relacionamento (`Relationship_Status`).

```python
# Ordena os grupos de Relationship_Status pela m√©dia de conflitos
order = df.groupby('Relationship_Status')['Conflicts_Over_Social_Media'].mean().sort_values(ascending=False).index

# Plot
plt.figure(figsize=(10,6))
sns.boxplot(x='Relationship_Status', y='Conflicts_Over_Social_Media', data=df, order=order, palette='coolwarm')

plt.title('H3 ‚Äì Conflitos Sociais por Status de Relacionamento', fontsize=16, fontweight='bold')
plt.xlabel('Status de Relacionamento', fontsize=12)
plt.ylabel('N¬∫ de Conflitos Relacionados √†s Redes', fontsize=12)

plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()
```

### Hip√≥tese:

### Objetivo da An√°lise:
Embora o dataset n√£o contenha uma vari√°vel expl√≠cita que me√ßa diretamente a preocupa√ß√£o dos usu√°rios com valida√ß√£o social (como n√∫mero de curtidas, seguidores ou tempo em plataformas espec√≠ficas), √© poss√≠vel utilizar a vari√°vel `Most_Used_Platform` como um **indicador indireto**.

```python
# Filtra apenas plataformas voltadas para valida√ß√£o social
validacao_social = ['Instagram', 'TikTok', 'Facebook', 'Snapchat']
df_validacao = df[df['Most_Used_Platform'].isin(validacao_social)]

# Ordena pela m√©dia de conflitos
order = df_validacao.groupby('Most_Used_Platform')['Conflicts_Over_Social_Media'].mean().sort_values(ascending=False).index

# Plot refinado
plt.figure(figsize=(8,5))
sns.barplot(data=df_validacao, x='Most_Used_Platform', y='Conflicts_Over_Social_Media', order=order, ci=None, palette='muted')

plt.title('H4 ‚Äì Conflitos em Plataformas com Valida√ß√£o Social', fontsize=16, fontweight='bold')
plt.xlabel('Plataforma')
plt.ylabel('M√©dia de Conflitos')
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

A an√°lise revela que os estudantes que utilizam com maior frequ√™ncia plataformas como **Instagram** e **TikTok** apresentam, em m√©dia, **um n√∫mero maior de conflitos sociais relacionados ao uso das redes**.
