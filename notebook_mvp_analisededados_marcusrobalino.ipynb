## 1. Descrição do Problema  

Com a crescente digitalização das interações sociais, o uso excessivo de redes sociais entre jovens adultos e adolescentes tem gerado preocupações sobre os impactos na saúde mental e nos relacionamentos interpessoais. Este trabalho busca compreender como esse comportamento influencia aspectos como foco, qualidade do sono, ansiedade social e conflitos nas relações sociais e afetivas.  

O objetivo central é realizar uma análise exploratória para identificar padrões e possíveis correlações entre o nível de dependência digital e os impactos percebidos pelos próprios estudantes.  

---

## 2. Tipo de Problema  

Este é um problema de **análise exploratória de dados**, classificado como **não supervisionado**, já que não há uma variável-alvo específica a ser prevista. O foco está na investigação de padrões, correlações e insights extraídos a partir dos dados disponíveis.

---

## 3. Premissas Gerais  

A hipótese de partida é que o uso intensivo de redes sociais modifica a forma como os jovens se relacionam com o mundo ao seu redor. A exposição constante a conteúdos altamente personalizados, reforçada por algoritmos, pode criar bolhas de afinidade e distorcer a percepção da realidade social.  

Esses fatores podem resultar em:  
- Dificuldades em lidar com opiniões divergentes  
- Redução da qualidade das interações presenciais  
- Problemas de saúde mental relacionados a comparação social, ansiedade e baixa autoestima  

---

## 4. Hipóteses Específicas  

**H1 –** Estudantes com maior tempo diário de uso de redes sociais apresentam níveis mais elevados de ansiedade social e maior dificuldade de foco em interações presenciais.  

**H2 –** Altos níveis de ansiedade social e baixa qualidade de sono estão correlacionados com maiores scores de vício em redes sociais.  

**H3 –** Estudantes em relacionamentos afetivos estáveis (status "In Relationship" ou "Complicated") relatam mais conflitos relacionados ao uso de redes sociais do que estudantes solteiros.  

**H4 –** Usuários que demonstram maior preocupação com validação social (ex.: número de curtidas e comentários) tendem a relatar também mais conflitos interpessoais gerados pelo uso de redes sociais.  

**H5 –** Estudantes que relatam maior número de conflitos interpessoais tendem a apresentar níveis mais altos de vício digital, sugerindo um ciclo de reforço negativo entre uso excessivo e insatisfação nas relações sociais.

###📌 3. Transformação de Variáveis Numéricas

### Normalização e Padronização de Variáveis Numéricas:

Nesta etapa, aplicamos duas técnicas de transformação em três variáveis numéricas do dataset:

- **Variáveis transformadas:**  
`Avg_Daily_Usage_Hours`, `Sleep_Hours_Per_Night`, `Mental_Health_Score`

---

#### ✔️ StandardScaler (Padronização):
Transforma os dados para uma escala com média 0 e desvio padrão 1.  
**Objetivo:** Reduzir os efeitos de variáveis com magnitudes muito diferentes, preservando as distribuições relativas.

**Variáveis padronizadas:**  
- `Avg_Daily_Usage_Hours_StandardScaled`  
- `Sleep_Hours_Per_Night_StandardScaled`  
- `Mental_Health_Score_StandardScaled`

---

#### ✔️ MinMaxScaler (Normalização):
Transforma os dados para um intervalo fixo (0 a 1), preservando a proporcionalidade entre os valores originais.  
**Objetivo:** Tornar os dados comparáveis em escalas fixas.

**Variáveis normalizadas:**  
- `Avg_Daily_Usage_Hours_MinMaxScaled`  
- `Sleep_Hours_Per_Night_MinMaxScaled`  
- `Mental_Health_Score_MinMaxScaled`

---

Ao final deste processo, o dataset conta agora com **seis novas colunas**, prontas para futuras análises quantitativas mais avançadas.

Essa preparação é fundamental para garantir a qualidade e a integridade dos resultados nas próximas etapas.


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Criando cópias das colunas numéricas que vamos transformar
numerical_features = ['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score']

# Aplicando StandardScaler

scaler_standard = StandardScaler()
scaled_standard = scaler_standard.fit_transform(df[numerical_features])

# Criando um DataFrame com os resultados do StandardScaler
df_standard = pd.DataFrame(scaled_standard, columns=[col + '_StandardScaled' for col in numerical_features])

# Concatenando ao DataFrame original
df = pd.concat([df, df_standard], axis=1)

# Aplicando MinMaxScaler

scaler_minmax = MinMaxScaler()
scaled_minmax = scaler_minmax.fit_transform(df[numerical_features])

# Criando um DataFrame com os resultados do MinMaxScaler
df_minmax = pd.DataFrame(scaled_minmax, columns=[col + '_MinMaxScaled' for col in numerical_features])

# Concatenando ao DataFrame original
df = pd.concat([df, df_minmax], axis=1)

# Exibindo o resultado final

df.head()
```

**Objetivo da Análise:**  
Verificar a relação entre o tempo médio diário nas redes sociais (`Avg_Daily_Usage_Hours`) e o impacto na saúde mental (`Mental_Health_Score`), que é usado aqui como proxy de sintomas de ansiedade social.


```python
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')

sns.regplot(
    x='Avg_Daily_Usage_Hours',
    y='Mental_Health_Score',
    data=df,
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'red'}
)

plt.title('H1 – Relação entre Tempo de Uso e Saúde Mental', fontsize=16, fontweight='bold')
plt.xlabel('Tempo Médio de Uso Diário (Horas)', fontsize=12)
plt.ylabel('Mental Health Score', fontsize=12)

# Grade horizontal mais leve
plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()
```

### Resultado da H1:

A análise gráfica revela uma tendência negativa: estudantes que passam mais tempo nas redes sociais apresentam, em média, **piores avaliações de saúde mental** (menor Mental Health Score).

Embora a dispersão dos dados seja considerável (característica comum em pesquisas comportamentais), a linha de regressão confirma a existência de um padrão de associação entre maior tempo de uso e pior percepção subjetiva da saúde mental.

Essa evidência sustenta a hipótese inicial.

## Análise da Hipótese 2 (H2)

**Hipótese:**  
Ansiedade social e baixa qualidade de sono estão associadas a maiores escores de vício em redes sociais.

**Objetivo da Análise:**  
Investigar a relação entre três variáveis principais:  
- **Mental_Health_Score** (proxy para ansiedade social)  
- **Sleep_Hours_Per_Night** (qualidade de sono)  
- **Addicted_Score** (nível de vício digital)

O objetivo é verificar se estudantes com pior avaliação de saúde mental e menor quantidade de sono apresentam níveis mais altos de adição às redes sociais.


```python
sns.set_style('white')

plt.figure(figsize=(8,6))
sns.heatmap(
    df[['Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Addicted_Score']].corr(),
    annot=True,
    cmap='YlOrRd',
    fmt=".2f",
    linewidths=0.5,
    linecolor='white',
    cbar_kws={'shrink': 0.7}
)

plt.title('H2 – Correlação: Sono, Saúde Mental e Nível de Vício', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
df['Faixa_Sono'] = pd.cut(df['Sleep_Hours_Per_Night'], bins=[0,5,7,10], labels=['Baixo Sono','Sono Médio','Alto Sono'])

# Boxplot visualmente refinado
plt.figure(figsize=(8,6))
sns.boxplot(x='Faixa_Sono', y='Addicted_Score', data=df, palette='Oranges')

plt.title('H2 – Nível de Vício por Faixa de Sono', fontsize=16, fontweight='bold')
plt.xlabel('Faixa de Sono', fontsize=12)
plt.ylabel('Addicted Score', fontsize=12)

plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

### Resultado da H2:

O heatmap evidencia duas relações importantes:

- Uma **correlação negativa entre horas de sono e Addicted_Score**, sugerindo que estudantes com menor tempo de sono tendem a apresentar níveis mais altos de adição digital.

- Uma **correlação fortemente negativa entre Mental_Health_Score e Addicted_Score**, indicando que estudantes com pior avaliação de saúde mental também relatam níveis mais elevados de vício nas redes sociais.

Além disso, o boxplot por faixas de sono reforça que os estudantes com **menos de 5 horas de sono por noite** (faixa "Baixo Sono") apresentam as **maiores medianas de Addicted_Score**, sustentando a hipótese de que **baixa qualidade de sono está associada a maior dependência digital**.

Esses padrões fortalecem a hipótese de que fatores como **ansiedade social e privação de sono** estão diretamente ligados ao **aumento do comportamento aditivo em redes sociais**.

## Análise da Hipótese 3 (H3)

**Hipótese:**  
Estudantes em relacionamentos afetivos estáveis apresentam mais conflitos relacionados ao uso de redes sociais do que estudantes solteiros.

**Objetivo da Análise:**  
Verificar se existe diferença significativa no número de conflitos sociais gerados pelo uso das redes sociais (`Conflicts_Over_Social_Media`) entre estudantes com diferentes status de relacionamento (`Relationship_Status`).

O foco é comparar visualmente a distribuição de conflitos entre os grupos: **Single**, **In Relationship** e **Complicated**.


```python
# Ordena os grupos de Relationship_Status pela média de conflitos
order = df.groupby('Relationship_Status')['Conflicts_Over_Social_Media'].mean().sort_values(ascending=False).index

# Plot
plt.figure(figsize=(10,6))
sns.boxplot(x='Relationship_Status', y='Conflicts_Over_Social_Media', data=df, order=order, palette='coolwarm')

plt.title('H3 – Conflitos Sociais por Status de Relacionamento', fontsize=16, fontweight='bold')
plt.xlabel('Status de Relacionamento', fontsize=12)
plt.ylabel('Nº de Conflitos Relacionados às Redes', fontsize=12)

plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()
```

### Resultado da H3:

O boxplot mostra que os estudantes classificados como **"In Relationship"** e **"Complicated"** apresentam uma distribuição de conflitos sociais **ligeiramente mais concentrada nas faixas superiores**, quando comparados ao grupo **"Single"**.

Embora as **medianas estejam próximas nos três grupos**, observa-se uma **maior densidade de valores altos** entre os estudantes com algum tipo de vínculo afetivo.

Esses padrões sugerem que a **vida relacional pode estar associada a mais situações de tensão social relacionadas ao uso das redes**, o que dá suporte à hipótese de que relacionamentos afetivos intensificam os impactos sociais do comportamento digital.

A evidência é sutil, mas coerente com a hipótese proposta.


## Análise da Hipótese 4 (H4)

### Hipótese:
Usuários mais preocupados com validação social tendem a relatar mais conflitos interpessoais.

### Objetivo da Análise:
Embora o dataset não contenha uma variável explícita que meça diretamente a preocupação dos usuários com validação social (como número de curtidas, seguidores ou tempo em plataformas específicas), é possível utilizar a variável `Most_Used_Platform` como um **indicador indireto**.

Para manter a coerência com a hipótese, a análise foi direcionada exclusivamente para plataformas com forte apelo à validação social pública — ou seja, redes que incentivam comparações sociais visíveis por meio de curtidas, comentários e seguidores. Foram incluídas: **Instagram, TikTok, Facebook e Snapchat**.

O objetivo é verificar se estudantes que utilizam majoritariamente essas plataformas apresentam, em média, **um número maior de conflitos sociais relacionados ao uso das redes** (`Conflicts_Over_Social_Media`), em comparação com os demais.

Essa abordagem permite manter o foco da hipótese sem distorcer a leitura dos dados, concentrando a análise nas redes que mais se conectam ao comportamento de validação social.


```python
# Filtra apenas plataformas voltadas para validação social
validacao_social = ['Instagram', 'TikTok', 'Facebook', 'Snapchat']
df_validacao = df[df['Most_Used_Platform'].isin(validacao_social)]

# Ordena pela média de conflitos
order = df_validacao.groupby('Most_Used_Platform')['Conflicts_Over_Social_Media'].mean().sort_values(ascending=False).index

# Plot refinado
plt.figure(figsize=(8,5))
sns.barplot(data=df_validacao, x='Most_Used_Platform', y='Conflicts_Over_Social_Media', order=order, ci=None, palette='muted')

plt.title('H4 – Conflitos em Plataformas com Validação Social', fontsize=16, fontweight='bold')
plt.xlabel('Plataforma')
plt.ylabel('Média de Conflitos')
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

### Resultado da H4:

A análise revela que os estudantes que utilizam com maior frequência plataformas como **Instagram** e **TikTok** apresentam, em média, **um número maior de conflitos sociais relacionados ao uso das redes**.

Esses resultados são coerentes com a hipótese, considerando que essas plataformas possuem mecanismos de engajamento fortemente baseados em **validação social pública** (número de curtidas, seguidores, comentários, etc).

Embora o dataset não permita uma análise direta sobre a preocupação emocional dos usuários com a validação social, a **correlação indireta entre o tipo de plataforma e o número de conflitos** sustenta a hipótese proposta.

Essa evidência reforça a necessidade de estudos futuros com métricas mais objetivas sobre comportamento de validação digital.


## Análise da Hipótese 5 (H5)

**Hipótese:**  
Um maior número de conflitos sociais pode reforçar o ciclo de dependência das redes sociais.

**Objetivo da Análise:**  
Investigar a existência de uma relação direta entre o número de conflitos sociais relatados pelos estudantes (`Conflicts_Over_Social_Media`) e o nível de adição digital (`Addicted_Score`).  

A ideia é verificar se **estudantes que relatam mais conflitos também apresentam maiores níveis de vício**, sugerindo um possível ciclo vicioso de uso excessivo das redes sociais como mecanismo de fuga ou compensação emocional.


```python
sns.set_style('whitegrid')

plt.figure(figsize=(10,6))
sns.regplot(
    x='Conflicts_Over_Social_Media',
    y='Addicted_Score',
    data=df,
    scatter_kws={'alpha': 0.6},
    line_kws={"color": "red"}
)

plt.title('H5 – Relação entre Nº de Conflitos e Nível de Vício Digital', fontsize=16, fontweight='bold')
plt.xlabel('Nº de Conflitos Sociais Relacionados às Redes', fontsize=12)
plt.ylabel('Addicted Score', fontsize=12)

plt.grid(axis='y', color='lightgray', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

### Resultado da H5:

O gráfico de dispersão com linha de tendência revela uma **correlação linear positiva clara entre o número de conflitos sociais e o Addicted_Score**.  

Ou seja, **quanto maior o número de conflitos interpessoais relatados, maior tende a ser o nível de adição digital** entre os estudantes.

Esse padrão sugere a existência de uma possível **relação cíclica**, na qual os conflitos sociais relacionados ao uso das redes geram estresse ou desconforto emocional — levando os estudantes a buscar nas próprias redes uma forma de compensação, distração ou validação.  
Esse comportamento pode reforçar ainda mais o uso excessivo, contribuindo para a manutenção do vício.

Embora o gráfico não comprove causalidade, a força da correlação visualizada reforça a necessidade de estudos futuros com metodologia longitudinal, além de sugerir que **intervenções voltadas para a redução de conflitos interpessoais** podem ter impacto direto na diminuição do uso compulsivo das redes sociais.

Essa evidência sustenta a hipótese inicial.
